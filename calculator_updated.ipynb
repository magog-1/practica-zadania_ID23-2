```python
import math
def calc(expression):   
    
    numbers_dct = {
        'ноль': 0,
        'один': 1,
        'два': 2,
        'три': 3,
        'четыре': 4,
        'пять': 5,
        'шесть': 6,
        'семь': 7,
        'восемь': 8,
        'девять': 9,
        'десять': 10,
        'одиннадцать': 11,
        'двенадцать': 12,
        'тринадцать': 13,
        'четырнадцать': 14,
        'пятнадцать': 15,
        'шестнадцать': 16,
        'семнадцать': 17,
        'восемнадцать': 18,
        'девятнадцать': 19,
        'двадцать': 20,
        'тридцать': 30,
        'сорок': 40,
        'пятьдесят': 50,
        'шестьдесят': 60,
        'семьдесят': 70,
        'восемьдесят': 80,
        'девяносто': 90,
        'плюс': '+',
        'минус': '-',
        'умножить': '*',
        'разделить на': '',
        'остаток от': '',
        'разделить':'/', 
        'остаток':'%', 
        'на':'',
        'от':'',
        'десятых': 0.1,
        'сотых': 0.01, 
        'тысячных': 0.001,
        'степени': '**',
        'синус': math.sin, 
        'косинус': math.sin, 
        'тангенс':math.tan,
        'пи': math.pi
        }
    
    digit = 0
    operation = ''
    expression_list = expression.split()
    
    if 'в' in expression_list:
        expression_list.remove('в')
    if 'от' in expression_list:
        expression_list.remove('от')
    
    decimal_prt = []
    
    #print(expression_list)

    # Поиск десятичной части
    while True:
        if 'и' in expression_list:
            indx = expression_list.index('и')
            decimal_prt.append(numbers_dct[expression_list[indx + 1]] * numbers_dct[expression_list[indx + 2]])
            del expression_list[indx:indx+3]
        else:
            break

    # Поиск целой части
    for el in expression_list:
        if type(numbers_dct[el]) is int:
            digit += numbers_dct[el]
        else:
            if digit != 0:                              
                digit_1 = digit
            if operation:
                operation += ' '
            operation += el
            digit = 0
    
    if len(decimal_prt) > 0:
        if len(decimal_prt) == 1:
            digit_1 += decimal_prt[0]
        else:
            digit_1 += decimal_prt[0]
            digit += decimal_prt[1]
    
    
    # Операции
    operation = numbers_dct[operation]
    if operation == '+':
        digit_1 += digit 
    elif operation == '-':
        digit_1 -= digit 
    elif operation == '*':
        digit_1 *= digit 
    elif operation == '/':
        digit_1 /= digit
    elif operation == '%':
        digit_1 %= digit
    elif operation == '**':
        digit_1 **= digit
    elif operation == math.sin:
        digit_1 = digit
        digit_1 = math.sin(digit_1)
    elif operation == math.cos:
        digit_1 = digit
        digit_1 = math.cos(digit_1)
    elif operation == math.tan:
        digit_1 = digit
        digit_1 = math.tan(digit_1)
    #print(digit_1)
    
   # Перевод числа в текстовую запись 
    def number_to_words(digit_1):
        units = ['', 'один', 'два', 'три', 'четыре', 'пять', 'шесть', 'семь', 'восемь', 'девять']
        tens = ['', '', 'двадцать', 'тридцать', 'сорок', 'пятьдесят', 'шестьдесят', 'семьдесят', 'восемьдесят', 'девяносто']
        teens = ['десять', 'одиннадцать', 'двенадцать', 'тринадцать', 'четырнадцать', 'пятнадцать', 'шестнадцать', 'семнадцать', 'восемнадцать', 'девятнадцать']
        hundreds = ['', 'сто', 'двести', 'триста', 'четыреста', 'пятьсот', 'шестьсот', 'семьсот', 'восемьсот', 'девятьсот']

        if digit_1 == 0:
            return 'ноль'

        if digit_1 < 0:
            return 'минус ' + number_to_words(abs(digit_1))

        result = ''

        integer_part = int(digit_1)
        decimal_part = int(round((digit_1 - integer_part) * 10))

        if integer_part > 0:
            if (integer_part // 100) > 0:
                result += hundreds[integer_part // 100] + ' '
                integer_part %= 100

            if (integer_part // 10) > 1:
                result += tens[integer_part // 10] + ' '
                integer_part %= 10

            if integer_part == 0:
                result += 'целых'
            elif integer_part < 10:
                result += units[integer_part]
            else:
                result += teens[integer_part - 10]

        if decimal_part > 0:
            result += ' '
            if integer_part > 0:
                result += 'и '

            if (decimal_part // 10) > 1:
                result += tens[decimal_part // 10] + ' '
                decimal_part %= 10

            if decimal_part < 10:
                result += units[decimal_part] + ' десятых'
            else:
                result += teens[decimal_part - 10] + ' десятых'
        

        
        return result.strip()
    return number_to_words(digit_1)


print(calc('двадцать пять плюс тринадцать'))
print(calc('девять разделить на три'))
print(calc('девять остаток от два'))
print(calc('пять и пять десятых плюс два'))
print(calc('пять минус шесть'))
print(calc('пять в степени два'))
print(calc('синус от один'))
print(calc('тангенс от два'))
```
