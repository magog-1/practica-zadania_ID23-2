```python
def calc(s):   
    
    dct = {
        'ноль': 0,
        'один': 1,
        'два': 2,
        'три': 3,
        'четыре': 4,
        'пять': 5,
        'шесть': 6,
        'семь': 7,
        'восемь': 8,
        'девять': 9,
        'десять': 10,
        'одиннадцать': 11,
        'двенадцать': 12,
        'тринадцать': 13,
        'четырнадцать': 14,
        'пятнадцать': 15,
        'шестнадцать': 16,
        'семнадцать': 17,
        'восемнадцать': 18,
        'девятнадцать': 19,
        'двадцать': 20,
        'тридцать': 30,
        'сорок': 40,
        'пятьдесят': 50,
        'шестьдесят': 60,
        'семьдесят': 70,
        'восемьдесят': 80,
        'девяносто': 90,
        'плюс': '+',
        'минус': '-',
        'умножить': '*'
    }
    

    d = 0
    
    for el in s.split():
        if type(dct[el]) == int:
            d += dct[el]
        else:
            d1 = d
            oper = dct[el]
            d = 0
    
    if oper == '+':
        d1 += d 
    elif oper == '-':
        d1 -= d 
    else:
        d1 *= d 
    
    dct_res = {v: k for k, v in dct.items()}
    result = []
    
    if d1 < 0:
        result.append('минус')
        d1 *= -1
    elif d1 == 0:
        result.append('ноль')
    
    k = 1000
    
    while k:
        if d1 > k-1:
            if d1 > 20:
                result.append(dct_res[d1//k*k])
                d1 %= k 
            else:
                result.append(dct_res[d1])
                k = 0
        k //= 10
    
    return ' '.join(result)
    

print(calc(input()))   #Пример: двадцать пять плюс тринадцать
```
