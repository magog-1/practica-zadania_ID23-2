```python
#1
class DynamicArray(object):
    def __init__(self):
        self.array = []

    def append(self, e):
        self.array.append(e)

    def get(self, index):
        return self.array[index]

    def set(self, index, value):
        self.array[index] = value

    def length(self):
        return len(self.array)

    def remove(self, index):
        del self.array[index]

    def __str__(self):
        return str(self.array)

arr = DynamicArray()
arr.append(1)
arr.append(2)
arr.append(3)

print(arr)
print(arr.length())

arr.remove(1)
print(arr)

arr.set(0, 5)
print(arr)
print(arr.get(1))
print(arr.__str__(), type(arr.__str__()))
```
